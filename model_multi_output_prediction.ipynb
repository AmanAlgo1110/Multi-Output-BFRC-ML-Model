```python
outliers_columns=["Water-cement ratio","Crude aggregate kg/m³","Fine aggregate kg/m³","Silica ash kg/m³","Fiber diameter-mm","Fiber content-%"]
```

```python
outlier_feature=data[outliers_columns]
```

```python
outlier_feature
```

```python
from sklearn.neighbors import LocalOutlierFactor
lf=LocalOutlierFactor(n_neighbors=20,contamination=.1)
y_pred=lf.fit_predict(outlier_feature)
```

```python
y_pred
```

```python
outlier_feature_outliers=outlier_feature[y_pred==-1]
```

```python
index_outlier_feature=outlier_feature_outliers.index
```

```python
data_modified=data.copy()
```

```python
for i in data_modified.columns:
    if(i in outliers_columns):
        for j in data_modified[i].index:
            if(j in index_outlier_feature):
                data_modified[i][j]=np.nan
```

```python
data_modified.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=8)
imputed_data=imputer.fit_transform(data_modified[outliers_columns])
data_modified[outliers_columns] = pd.DataFrame(imputed_data, columns=outliers_columns, index=data_modified.index)
```
